# Enrichissement des données foncières avec la population communale

```python
import pandas as pd

data = pd.read_csv("../clean_data/ValeursFoncieres_nettoyees.csv",
                   dtype={
                       "Code commune": str,
                        "No plan": str,
                        "Code type local": str,
                        "Code departement": str
                        })
```

```python
data['CodeINSEE'] = data['Code departement'] + data['Code commune']
```

```python
pop = pd.read_csv("../data/donnees_communes.csv", sep=";", encoding="utf-8")
```

```python
pop['CodeINSEE'] = pop['COM'].astype(str).str.zfill(5)

data = data.merge(pop[['CodeINSEE', 'PTOT']], on='CodeINSEE', how='left')

data.rename(columns={'PTOT': 'Population'}, inplace=True)

data.head()
```

```python
densite =  pd.read_excel("../data/densite_commune.xlsx")
```

```python
data = pd.merge(data, densite[['codgeo', 'dens_pop']], left_on='CodeINSEE', right_on='codgeo', how='left')
data.drop(columns=['codgeo'], inplace=True)
data.head(5)
```

```python
nom_departement = pd.read_excel("../data/nom_departement.xlsx")
data = pd.merge(data, nom_departement, left_on='Code departement', right_on='DEP', how='left')
data.drop(columns='DEP', inplace=True)
data.head(5)
```

```python
data.to_csv("../clean_data/ValeursFoncieres_enrichies.csv", index=False, encoding="utf-8")
```
